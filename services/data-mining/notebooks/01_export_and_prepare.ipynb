## Data mining: export from MongoDB and prepare dataset


```python
import os

MONGODB_URI = os.environ.get("MONGODB_URI", "")
MONGODB_DB = os.environ.get("MONGODB_DB", "")
(MONGODB_URI[:40] + "...", MONGODB_DB)
```

```python
from pathlib import Path

raw_dir = Path("services/data-mining/data/raw")
processed_dir = Path("services/data-mining/data/processed")
raw_dir.mkdir(parents=True, exist_ok=True)
processed_dir.mkdir(parents=True, exist_ok=True)
(raw_dir, processed_dir)
```

```python
from export_mongo import export

if not MONGODB_URI:
    raise ValueError("Set MONGODB_URI first")

export(
    mongo_uri=MONGODB_URI,
    db_name=MONGODB_DB or None,
    out_dir=str(raw_dir),
    users_collection="users",
    products_collection="products",
    orders_collection="orders",
    limit=2000,
)
```

```python
from prepare_propensity_dataset import build_dataset

build_dataset(
    users_csv=str(raw_dir / "users.csv"),
    products_csv=str(raw_dir / "products.csv"),
    orders_csv=str(raw_dir / "orders.csv"),
    order_items_csv=str(raw_dir / "order_items.csv"),
    out_csv=str(processed_dir / "propensity_dataset.csv"),
    label_window_days=30,
    min_history_orders=2,
)
```
